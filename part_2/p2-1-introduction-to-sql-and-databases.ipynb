# Part 2.1: Introduction to SQL and Databases

Importing sqlite3 to run queries in python:

```python
import pandas as pd
import sqlite3
```

Connecting to the database and creating a cursor object:

```python
connection = sqlite3.connect('chinook.db')
cursor = connection.cursor()
```

## Introduction to SQL

A simple SELECT:

Making the query and then reading the SQL query into a pandas dataframe:

```python
query = "SELECT 2+3;"

pd.read_sql_query(query,connection)
```

Multiple calculations:

```python
query = "SELECT 1+3+5, 1+10, 2+4+6+8+10;"

pd.read_sql_query(query,connection)
```

Naming results:

```python
query = "SELECT 2+3+5+7 AS sum_first_prime_numbers;"

pd.read_sql_query(query,connection)
```

Arithmetic in SQL and using cleaner code:

```python
query = """

SELECT
    2+3+5 AS sum_prime,
    20-3 AS sub,
    16/2 AS div,
    3*5 AS mult,
    3+5*2 AS sum_mult;
    
"""

pd.read_sql_query(query,connection)
```

Removing the cursor object and closing the connection to the database:

```python
cursor.close()
connection.close()
```

## Introduction to Databases

Connecting to the database and creating a cursor object:

```python
connection = sqlite3.connect('chinook.db')
cursor = connection.cursor()
```

Extracting Data FROM Tables:

```python
query = """

SELECT
    first_name, 
    last_name, 
    birthdate, 
    phone, 
    email
FROM
    employee;
    
"""

pd.read_sql_query(query,connection)
```

Selecting All Columns

```python
query = """

SELECT
    *
FROM 
    employee;
    
"""

pd.read_sql_query(query,connection)
```

## Exploring Tables

Previewing Results:

```python
query = """

SELECT 
    *
FROM 
    invoice_line
LIMIT 
    10;
    
"""

pd.read_sql_query(query,connection)
```

Unique Rows:

```python
query = """

SELECT DISTINCT
    unit_price
FROM
    invoice_line
    
"""

pd.read_sql_query(query,connection)
```
